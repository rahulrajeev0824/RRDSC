```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
```

**Summary:** This data set from JPL is data about asteroids and their characteristics that have been overall classified as either hazardous and nonhazardous to impact on Earth. Potentially Hazardous Asteroids or PHAs are classfied based on parameters that measure it's potential to make dangerously close approaches to the Earth. All asteroids with an Earth Minimum Orbit Intersection Distance (MOID) of 0.05 AU (astronomical units) or less and an absolute magnitude (unitless) of 22 or less are considered PHAs. Asteroids that don't fit under these conditions and smaller than 140 m in diameter are not considered PHAs. 

**Source:** https://www.kaggle.com/datasets/shrushtijoshi/asteroid-impacts

```python
# loading the 
orbit_df = pd.read_csv('orbits.csv')
```

**2. Identify a question or question(s) that you would like to explore in your data set.**

I want to see whether MOID and Absolute Magnitude are collinear, and if not, whether they can be used in a predictive model.
I will have to plot two histograms and a scatter plot comparing the two.


```python
# creating subsets of data
magnitudes = orbit_df['Asteroid Magnitude']
moid = orbit_df['Minimum Orbit Intersection Distance (AU)']
```

**3. Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are labeled appropriately and professionally.**

```python
# Magnitude Histogram

# figure size
plt.figure(figsize = (12,8))

# histogram function
plt.hist(magnitudes, bins = [*range(10,40,1)], color = 'lightblue', edgecolor = 'black', linewidth = 1)

#creating ticks that work for the set of values we have
xticks = [*range(10,40,1)]
plt.xticks(ticks=xticks)

# labels
plt.xlabel('Magnitude (unitless)')
plt.ylabel('Frequency')

# show figure
plt.show()
```

```python
# MOID Histogram

# figure size
plt.figure(figsize = (12,8))

# histogram function with appropriate bins to contain main distribution
plt.hist(moid, bins = np.linspace(0,0.8,8, endpoint=False), color = 'crimson', edgecolor = 'black', linewidth = 1)

# tick markers and limiting the distribution range to show data
plt.xticks(np.linspace(0,0.8,8, endpoint=False))

# labels
plt.xlabel('Moid (AU)')
plt.ylabel('Frequency')
```

```python
# creating scatter plot
plt.scatter(magnitudes, moid, alpha=1.0, s=5)
plt.xlabel('Magnitude (unitless)')
plt.ylabel('MOID (AU)')
```

**4. Explain what you have learned from each of your graphs.**

The magnitude histogram is relatively uniformly distributed. The MOID histogram is slightly skew left, but that is normal as MOIDs past this range aren't typically in any danger of colliding with the Earth any time soon. The asteroids that are detected as PHAs are within this range anyways. From the scatter plot, I found that there is no linear correlation between hte magnitude and the MOID.
