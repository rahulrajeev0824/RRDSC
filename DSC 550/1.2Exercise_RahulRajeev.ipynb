```python
# import
import pandas as pd
from collections import Counter
import numpy as np
```

```python
# 1. Load data set into pandas
vgsales = pd.read_csv('videogamessales2016.csv', delimiter=',')
```

```python
# 2. Display the first 10 rows
vgsales.head(10)
```

```python
# 3. dimensions of the dataset, and explain the context

# use shape to get out the rows and columns tuple
nrow, ncol = vgsales.shape
print(nrow, ncol)
```

**Thoughts:** In the context of this dataset, there are 16 different attributes for video games, and there are a total of 16719 games.

```python
# 4. Top 5 games by critic score

# sort_values with ascending parameter to be false
topcritic = vgsales.sort_values(by=['Critic_Score'], ascending = False)
topcritic.head()
```

```python
# 5. Find the number of video games in each genre

# create a list of genre entries from dataframe
genre = list(vgsales['Genre'])

# use counter to display the genres and counts as a dictionary
genrecounts = Counter(genre)

# revert to dataframe for easy viewing
genredf = pd.DataFrame.from_dict(genrecounts, orient='index')
genredf
```

```python
# 6. Find the first five games in the dataframe on the SNES platform

# creating subset
snesvg = vgsales[vgsales['Platform'] == 'SNES']
snesvg.head()

# sort by with ascending = True, then finding the first 5
sortedsnes = snesvg.sort_values(by='Year_of_Release')
sortedsnes.head()
```

```python
# 7. Find the five publishers with the highest total global sales. 
# Note: You will need to calculate the total global sales for each publisher to do this.

# group by publisher, sum the global sales column and reset the index to keep as dataframe
global_sales_by_pub = vgsales.groupby(by='Publisher')['Global_Sales'].sum().reset_index()

# sorting the values by global sales, ascending is false, and displaying first 5
global_sales_by_pub = global_sales_by_pub.sort_values(by = 'Global_Sales', ascending=False).head(5)
global_sales_by_pub
```

```python
# 8. Create a new column in the data frame that calculates the percentage of global sales from North America. 
# Display the first five rows of the new data frame.

# performing the operation and equating it to a new column in the dataframe
vgsales['NA_Percent'] = vgsales['NA_Sales'] / vgsales['Global_Sales']
vgsales.head()
```

```python
# 9. Find the number NaN entries (missing data values) in each column.
vgsales.isnull().sum().reset_index()
```

```python
# 10. Try to calculate the median user score of all the video games. 
# You will likely run into an error because some of the user score entries are 
# a non-numerical string that cannot be converted to a float. 
# Find and replace this string with NaN and then calculate the median. 
# Then, replace all NaN entries in the user score column with the median value.

# setting column of user scores to a variable
user_score = vgsales['User_Score']

# cleaning everything but a number using regex
user_score_clean = user_score.replace(r'^([A-Za-z]|[0-9]|_)+$', np.NaN, regex=True)

# demonstrating that the NaN count has increased from strings being replaced
print('Before:', user_score.isnull().sum(), 'After:',user_score_clean.isnull().sum())
```

```python
# replacing all the NaN values with the median of the cleaned column
user_score_clean = user_score.replace(np.NaN, user_score_clean.median())

# double checking the count after
print(user_score_clean.isnull().sum())
```

```python
# replacing the existing user score column with the cleaned version
vgsales['User_Score'] = user_score_clean
vgsales.head()
```

**Thoughts:** Here, to prove that the code did indeed work, I double checked with an earlier call to head() on the dataframe before the strings and NaN were replaced with median, and found that the Super Mario Bros on the NES now has a user_score of 7.5, the median of the user_scores.
